```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
boston_df = pd.read_csv('Boston.csv')
```

```python
# Create the boxplot
plt.boxplot(boston_df['medv'])

# Add title and labels
plt.title('Boxplot of Median Value of Owner-Occupied Homes')
plt.ylabel('Median Value ($1000s)')

# Show the plot
plt.show()
```

```python
# Calculate the median value of houses bounded by the Charles river
charles_river_median = boston_df[boston_df['chas'] == 1]['medv'].median()

# Calculate the median value of houses not bounded by the Charles river
no_charles_river_median = boston_df[boston_df['chas'] == 0]['medv'].median()

# Create a bar plot of the median values
plt.bar(['Charles River', 'No Charles River'], [charles_river_median, no_charles_river_median])
plt.title('Median Value of Houses by Charles River')
plt.xlabel('Bounded by Charles River')
plt.ylabel('Median Value of Houses ($1000s)')
plt.show()
```

Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

*Hypothesis:*

H0: The median value of houses bounded by the Charles river is not significantly different from the median value of houses not bounded by the Charles river.

H1: The median value of houses bounded by the Charles river is significantly different from the median value of houses not bounded by the Charles river.

We will use a two-sample t-test for independent samples to test this hypothesis.

```python
# Separate the MEDV variable by the CHAS variable (0 or 1)
chas_0 = boston_df[boston_df['chas'] == 0]['medv']
chas_1 = boston_df[boston_df['chas'] == 1]['medv']

# Perform the t-test
t_stat, p_val = stats.ttest_ind(chas_0, chas_1)

# Print the results
print('T-Statistic:', t_stat)
print('P-Value:', p_val)

# Check if the p-value is less than the significance level (alpha)
alpha = 0.05
if p_val < alpha:
    print('There is a significant difference in median value of houses bounded by the Charles river.')
else:
    print('There is no significant difference in median value of houses bounded by the Charles river.')
```

Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)



```python
# Create three groups based on the AGE variable
age_1 = boston_df[boston_df['age'] <= 35]['medv']
age_2 = boston_df[(boston_df['age'] > 35) & (boston_df['age'] <= 70)]['medv']
age_3 = boston_df[boston_df['age'] > 70]['medv']

# Perform ANOVA
f_stat, p_value = stats.f_oneway(age_1, age_2, age_3)

# Print the results
print("F-Statistic: ", f_stat)
print("P-Value: ", p_value)
```

The obtained F-statistic is 36.41 and the corresponding p-value is very small (1.71e-15), which suggests strong evidence against the null hypothesis. Therefore, we can conclude that there is a significant difference in the median values of houses for each proportion of owner-occupied units built prior to 1940 (AGE).

Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

```python
# calculating the correlation coefficient and p-value
corr, p_val = stats.pearsonr(boston_df['nox'], boston_df['indus'])

# printing the results
print("Correlation Coefficient:", corr)
print("P-Value:", p_val)

if p_val > 0.05:
    print("There is no significant relationship between Nitric oxide concentrations and proportion of non-retail business acres per town.")
else:
    print("There is a significant relationship between Nitric oxide concentrations and proportion of non-retail business acres per town.")
```
