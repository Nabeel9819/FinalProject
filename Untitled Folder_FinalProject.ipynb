```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
df = pd.read_csv('Boston.csv')

# Create the boxplot
plt.boxplot(df['medv'])
plt.title('Boxplot of MEDV Variable')
plt.ylabel('medv')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
df = pd.read_csv('Boston.csv')
river_counts = df['chas'].value_counts()
plt.bar(river_counts.index, river_counts.values)
plt.title('Barplot of CHAS Variable')
plt.xlabel('chas')
plt.ylabel('Counts')
plt.show()
```

```python
df.boxplot(column=['medv'], by='age')
plt.title('Boxplot of MEDV Variable vs AGE Variable')
plt.xlabel('age')
plt.ylabel('medv')
plt.show()
```

```python
plt.scatter(df['nox'], df['indus'])
plt.title('Scatter plot of NOX vs INDUS Variables')
plt.xlabel('nox')
plt.ylabel('indus')
plt.show()
```

```python
plt.hist(df['ptratio'], bins='auto')
plt.title('Histogram for Pupil-Teacher Ratio')
plt.xlabel('Pupil-Teacher Ratio')
plt.ylabel('Frequency')
plt.show()
```
